```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%matplotlib inline

# Custom color palette
custom_palette = ["#6C5CE7", "#00B894", "#FD79A8", "#FDCB6E"]

# Set theme properly
sns.set_theme(
    style="whitegrid",
    palette=custom_palette,
    rc={"figure.figsize": (12,6)},
    font_scale=1.3
)

# Load dataset
df = pd.read_csv(r"C:\PROJECT CHENNAI\rainfall.csv")

# Line plot
sns.lineplot(x='YEAR', y='JAN', data=df)

plt.title("Year vs January Rainfall")
plt.xlabel("Year")
plt.ylabel("January Rainfall")
plt.show()
```

```python
sns.histplot(df['ANNUAL'], kde=True)

plt.title("Distribution of Annual Rainfall")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv(r"C:\PROJECT CHENNAI\rainfall.csv")

sns.histplot(df['JAN'], bins=20, kde=False)

plt.title("Histogram of January Rainfall")
plt.xlabel("January Rainfall")
plt.ylabel("Frequency")
plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv(r"C:\PROJECT CHENNAI\rainfall.csv")

sns.kdeplot(y=df['JAN'], color='red', fill=False)

plt.title("Vertical KDE Plot of January Rainfall")
plt.ylabel("January Rainfall")
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv(r"C:\PROJECT CHENNAI\rainfall.csv")

# Select only few subdivisions (IMPORTANT STEP)
selected = df[df['SUBDIVISION'].isin(['Tamil Nadu', 'Kerala', 'Karnataka'])]

# Set theme
sns.set_theme(style="darkgrid", palette="crest")

# Pairplot
sns.pairplot(
    selected[['JAN','FEB','MAR','APR','SUBDIVISION']],
    hue='SUBDIVISION',
    diag_kind='kde'
)

plt.show()


```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\PROJECT CHENNAI\rainfall.csv")

sns.barplot(x='SUBDIVISION', y='JAN', data=df)

plt.xticks(rotation=90)
plt.title("Average January Rainfall by Subdivision")
plt.show()
```
