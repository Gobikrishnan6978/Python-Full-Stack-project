```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(r"C:\PROJECT CHENNAI\rainfall.csv")

plt.figure(figsize=(8,5))
plt.bar(df["YEAR"].head(10), df["JAN"].head(10))

plt.xlabel("Year")
plt.ylabel("January Rainfall")
plt.title("Year vs January Rainfall (First 10 Years)")

plt.xticks(rotation=45)    
plt.tight_layout()         
plt.show()
```

```python
plt.barh(YEAR, JAN)
plt.xlabel("Year")
plt.ylabel("January Rainfall")
plt.title("Year vs January Rainfall")
plt.show()
```

```python
YEAR = df['YEAR']
JAN = df['JAN']
plt.scatter(YEAR, JAN)
plt.title("Scatter Plot")
plt.xlabel("Year")
plt.ylabel("January Rainfall")
plt.title("Year vs January Rainfall")
plt.show()
```

```python
plt.pie(JAN.tail(5), labels=YEAR.tail(5))
plt.title("Pie Chart - Last 5 Years")
plt.show()
```
